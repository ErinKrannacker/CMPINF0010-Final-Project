I am going to determine the best neighborhood based on the number of playgrounds in each neighborhood. 

```python
#load pandas
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib

#load data from WPRDC
playgrounds = pd.read_csv("./Files/playgrounds.csv", index_col="id", parse_dates=True)
```

```python
#list how many playgrounds are in each neighborhood
p2 = playgrounds['neighborhood'].value_counts()
print(p2)
```

```python
#put into a bar plot
bPlot = p2.plot.bar(legend=False, figsize = (15,10))
plt.title("Playgrounds In Neighborhoods")
plt.xlabel("Neighborhoods") 
plt.ylabel("Number of Playgrounds")
```

```python
#narrow down to 3 or more
moreThanTwo = p2.loc[p2>2]
moreThanTwo.plot.bar(legend=False)
plt.title("Playgrounds In Neighborhoods")
plt.xlabel("Neighborhoods With Three or More Playgrounds")
plt.ylabel("Number of Playgrounds")
```

```python
#narrow down to 4 or more
moreThanThree = p2.loc[p2>3]
moreThanThree.plot.bar(legend=False)
plt.title("Playgrounds In Neighborhoods")
plt.xlabel("Neighborhoods With Four or More Playgrounds")
plt.ylabel("Number of Playgrounds")
```

From this, I determined that:
Squirrel Hill South had the most playgrounds, at 8. 
South Side Slopes and Beechview tied for second, both having 5. 
Allegheny Center, Beltzhoover, Sheraden, and Highland park are all tied at 3rd with 4 playgrounds.
